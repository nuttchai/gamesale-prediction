```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
videogame_data = pd.read_csv('videogame-sales.csv')
videogame_data
```

```python
df = videogame_data[['Platform', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].copy()
df
```

```python
# Get one hot encoding of columns 'Platform'
plat_onehot = pd.get_dummies(df['Platform'])
plat_onehot
```

```python
# Get one hot encoding of columns 'Publisher'
pub_onehot = pd.get_dummies(df['Publisher'])
pub_onehot
```

```python
# Get one hot encoding of columns 'Genre'
gen_onehot = pd.get_dummies(df['Genre'])
gen_onehot
```

```python
# Drop column as it is now encoded
data_one = df.copy()
data_one = data_one.drop('Platform',axis = 1)
data_one = data_one.drop('Genre',axis = 1)
data_one = data_one.drop('Publisher',axis = 1)

# Join the encoded data_one
data_one = data_one.join(plat_onehot)
data_one = data_one.join(gen_onehot)
data_one = data_one.join(pub_onehot, how='left', lsuffix='_left', rsuffix='_right')
data_one
```

```python
# Replacing 0 with 0.001 for NA_Sales
data_one.NA_Sales.replace({0: 0.001}, inplace=True)
```

### Create LR Model

#### construct X array and Y array

```python
X = data_one.iloc[:, 4:] # column-2600 to end
y_na = data_one.iloc[:,0:1] # Select NA_Sales column
```

```python
y_na
```

### split data into training set (80%) and test set (20%)


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_na, test_size=0.8, random_state=101)
```

```python
X_train.shape
```

```python
y_train.size
```

```python
X_test.shape
```

```python
y_test.size
```

### create and train the model


```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
X_train
```

```python
y_train
```

### evaluate the model

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_[0],X.columns,columns=['Coefficient'])
coeff_df
```

### use the model to predict the NA_Sales

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(y_test,predictions)
```

```python
lm.score(X_test, y_test)
```
